```python
import numpy as np
import matplotlib.pyplot as plt
import networkx as nx
```

`shuffle`, `normalize` are unused but can be investigated further

```python
N = 1000
t_max = 100000
f = np.empty((t_max, N)) # fitness value per species per timestep
f[0,:] = np.random.random(N)

def update(f, t=0, shuffle=False, normalize=False):
    f2 = f[t].copy()
    i = np.argmin(f2) # minimum
    i1 = (i - 1) % N # right neighbour
    i2 = (i + 1) % N # left neighbour
    if shuffle:
        indices = [j for j in range(N) if j not in [i, i1, i2]]
        f2 = list(f2[indices])
        for _ in range(3):
            # during this stage |f2| < N
            i = np.random.randint(0,len(f2))
            f2.insert(i, np.random.random())

        f2 = np.array(f2)

    else:
        f2[[i, i1, i2]] = np.random.random(3)

    if normalize:
        f2 = (f2 - np.min(f2)) * (np.max(f2) - np.min(f2))
    f[t+1,:] = f2

for t in range(t_max-1):
    update(f, t)
```

```python
# fig = plt.figure(figsize=(9,9))
fig = plt.figure(figsize=(5,9))
i_max = 100
plt.imshow(np.array(f)[:,:i_max], cmap='summer', aspect='auto')
plt.xlabel('species index')
plt.ylabel('time')
# plt.semilogy()
```

```python
N = 1000
t_max = 100000
f = np.empty((t_max, N)) # fitness value per species per timestep
f[0,:] = np.random.random(N)

def update(f, t=0):
    f2 = f[t].copy()
    i = np.argmin(f2) # minimum
    i1 = (i - 1) % N # right neighbour
    i2 = (i + 1) % N # left neighbour
    f2[[i, i1, i2]] = np.random.random(3) # update species
    f[t+1,:] = f2

for t in range(t_max-1):
    update(f, t)


# plot
fig = plt.figure(figsize=(5,9))
i_max = 100
plt.imshow(np.array(f)[:,:i_max], cmap='summer', aspect='auto')
plt.xlabel('species index')
plt.ylabel('time')
```

## Using networkx

```python
N = 100
t_max = 1000
f = np.empty((t_max, N)) # fitness value per species per timestep
f[0,:] = np.random.random(N)

# Create graph
g = nx.erdos_renyi_graph(N, 0.01)
#g = nx.watts_strogatz_graph(N, 3, 0.1)
#g = nx.barabasi_albert_graph(N, 1)

# perform Back-Sneppen
for t in range(t_max-1):
    i = np.argmin(f[t])
    idx = list(g[i])
    
    f[t+1] = f[t]
    f[t+1, i] = np.random.random()
    f[t+1, idx] = np.random.random(len(idx))
    
# plot
fig, ax = plt.subplots()
nx.draw_circular(g, with_labels=False)
ax.set_title("Network")

fig, ax = plt.subplots(figsize=(5,9))
ax.imshow(f[:,:], cmap="plasma", aspect="auto")
ax.set_xlabel("Species")
ax.set_ylabel("Time")
```

```python
N = 1000
t_max = 100000
f = np.empty((t_max, N))
age = np.empty((t_max, N))

f[0,:] = np.random.random(N)
age[0,:] = 0

# Create graph
#g = nx.erdos_renyi_graph(N, 0.01)
g = nx.watts_strogatz_graph(N, 3, 0)
#g = nx.barabasi_albert_graph(N, 1)
#g = nx.grid_graph([int(np.sqrt(N)), int(np.sqrt(N))], periodic=False)

# configure graph
# g = nx.convert_node_labels_to_integers(g)
# g = g.to_directed()

# variables for counting avalanches
min_fitness = -1
av_size = -1
avs = []

# perform Bak-Sneppen
for t in range(t_max-1):
    i = np.argmin(f[t])
    idx = list(g[i])
    
    if t == 0:
        min_fitness = f[t,i]
        av_size = 1
    elif f[t,i] < min_fitness:
        av_size += 1
    else:
        avs.append(av_size)
        min_fitness = f[t,i]
        av_size = 1
    
    f[t+1] = f[t]
    f[t+1, i] = np.random.random()
    f[t+1, idx] = np.random.random(len(idx))
    
    age[t+1] = age[t] + 1
    age[t+1, i] = 0
    age[t+1, idx] = 0
avs.append(av_size)
avs_count = np.unique(avs, return_counts=True)


# plot graph
fig, ax = plt.subplots()
nx.draw_circular(g, with_labels=False)
ax.set_title("Network")

# plot fitness and age
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 10))
fig.subplots_adjust(wspace=0.5)

im = ax1.imshow(f, cmap="plasma", aspect="auto")
ax1.set_xlabel("Species")
ax1.set_ylabel("Time")
cbar = plt.colorbar(im, ax=ax1)
cbar.set_label("Fitness")

im = ax2.imshow(age, cmap="viridis", aspect="auto")
ax2.set_xlabel("Species")
ax2.set_ylabel("Time")
cbar = plt.colorbar(im, ax=ax2)
cbar.set_label("Age")

# plot avalanche distribution
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12,5))
ax1.plot(avs, "o")
ax1.set_title("Avalanches")
ax1.set_xlabel("i")
ax1.set_ylabel("Avalanche size")

ax2.plot(*avs_count, "o")
ax2.set_xscale("log")
ax2.set_yscale("log")
ax2.set_title("Avalanche distribution")
ax2.set_xlabel("Avalanche size")
ax2.set_ylabel("Number of avalanches")

# plot fitness
fig, ax = plt.subplots()
ax.hist(np.min(f, axis=1), bins=20, density=True,
       histtype="step", label="Minimum fitness")
ax.hist(f.flatten(), bins=20, density=True,
       histtype="step", label="Fitness")
ax.set_xlabel("Fitness")
ax.set_ylabel("Frequency")
ax.legend(loc=9)
```
