```python
import numpy as np
import scipy
import matplotlib.pyplot as plt
import networkx as nx
np.random.seed(123)
```

`shuffle`, `normalize` are unused but can be investigated further

```python
N = 1000
t_max = 10000
f = np.empty((t_max, N)) # fitness value per species per timestep
f[0,:] = np.random.random(N)

def update(f, t=0, shuffle=False, normalize=False):
    f2 = f[t].copy()
    i = np.argmin(f2) # minimum
    i1 = (i - 1) % N # right neighbour
    i2 = (i + 1) % N # left neighbour
    if shuffle:
        indices = [j for j in range(N) if j not in [i, i1, i2]]
        f2 = list(f2[indices])
        for _ in range(3):
            # during this stage |f2| < N
            i = np.random.randint(0,len(f2))
            f2.insert(i, np.random.random())

        f2 = np.array(f2)

    else:
        f2[[i, i1, i2]] = np.random.random(3)

    if normalize:
        f2 = (f2 - np.min(f2)) * (np.max(f2) - np.min(f2))
    f[t+1,:] = f2

for t in range(t_max-1):
    update(f, t)
```

```python
fig = plt.figure(figsize=(9,15))
# fig = plt.figure(figsize=(5,9))
i_max = 100
plt.imshow(np.array(f)[:,:i_max], cmap='summer', aspect='auto')
plt.xlabel('species index')
plt.ylabel('time')
# plt.semilogy()
```

```python
N = 100
t_max = 1000
f = np.empty((t_max, N)) # fitness value per species per timestep
f[0,:] = np.random.random(N)

# Create graph
g = nx.erdos_renyi_graph(N, 0.01)
# g = nx.watts_strogatz_graph(N, 3, 0.1)
# g = nx.barabasi_albert_graph(N, 1)

# perform Back-Sneppen
for t in range(t_max-1):
    i = np.argmin(f[t])
    idx = list(g[i])
    
    f[t+1] = f[t]
    f[t+1, i] = np.random.random()
    f[t+1, idx] = np.random.random(len(idx))
    
# plot
fig, ax = plt.subplots()
nx.draw_circular(g, with_labels=False)
ax.set_title("Network")

fig, ax = plt.subplots(figsize=(5,9))
ax.imshow(f[:,:], cmap="plasma", aspect="auto")
ax.set_xlabel("Species")
ax.set_ylabel("Time")
```

```python
# class SpeciesPerPoint:
#     def __init__(self, identifiers=[], fitnesses=[]):
#         self.keys = identifiers
#         self.values = fitnesses

class Iterator:
    def __init__(self, start):
        self.next = start
        
    def get_next(self):
        self.next += 1
        return self.next - 1

N = 3
N2 = N**2
t_max = 3
N_species = 10
degree = 3
degree2 = np.ceil(degree/2).astype(int)
# N_edges = n_species * degree / 2
# assert n_edges <= scipy.special.binom(N_species, 2), 'L_\max'
# g = nx.barabasi_albert_graph(n_species, degree2, seed=123)
g = nx.erdos_renyi_graph(N_species, 0.01, seed=123, directed=False)
# g = nx.watts_strogatz_graph(N, 3, 0.1)
# g = nx.barabasi_albert_graph(N, 1)
grid = nx.lattice.grid_2d_graph(N, N, periodic=False)

# init species over spatial lattice
# note that the fitness is defined locally: it always depends on the environment
# f = np.empty((t_max, N2*N_species)) # fitness value per species per timestep
f = np.empty((t_max, N2), dtype=object) # fitness value per species per location per timestep
# f = [] # fitness value per species per location per timestep
# f[0,:] = np.random.random(f.shape[-1])
# f.append({k: v for k,v in range(N_species)})
for i in range(N2):
#     f[0,i] = {i:fitness for i,fitness in enumerate(np.random.random(N_species))}
    f[0,i] = FitnessArray(np.arange(N_species), np.random.random(N_species))

identifier_iter = Iterator(N_species)

# Back-Sneppen
for t in range(f.shape[0]-1):
# for t in [0]:
#     f[t+1] = f[t].copy()
    if t % 100 == 0:
        # TODO cleanup species from graph
        0

    for loc in range(f.shape[1]):
        # assume no interactions, therefore a single update stage
        i = np.argmin(f[t, loc])
        ids_extinctions = list(g[i]) + [i]
        f_array = f[t, loc]
        assert f_array is not None, (f_array, t, loc)
#         idx_extinctions_local = [k for k in f_array.keys if k not in idx_extinctions_global]
        idx_extinctions = [j for j in range(f_array.size) 
                                 if f_array.keys[j] in ids_extinctions]
        keys_new = [identifier_iter.get_next() for _ in range(len(idx_extinctions))]
        keys = np.concatenate([np.delete(f_array.keys, idx_extinctions),
                               keys_new
                              ])
        assert all([k in keys for k in keys_new])
        values_new = np.random.random(len(idx_extinctions))
        values = np.concatenate([np.delete(f_array, idx_extinctions),
                                 values_new
                                ])
        f[t+1, loc] = FitnessArray(keys, values)
        assert f[t+1, loc] is not None, 'post'
        g.add_nodes_from(keys_new)
        # TODO add random edges to new nodes
        # optional: only to locally available nodes
        # optional: use heuristic (e.g. BA)
        n_keys_old = keys.size - len(keys_new)
        for k1 in keys_new:
            # TODO make sure all k2 exist
            edges = [(k1, k2) for k2 in 
                     np.random.choice(keys[:n_keys_old], size=degree2)]
            g.add_edges_from(edges)
```
